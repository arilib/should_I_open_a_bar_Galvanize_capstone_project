```python
mpl.rcParams.update({
    'font.size'           : 14.0,
    'axes.titlesize'      : 'large',
    'axes.labelsize'      : 'large',
    'xtick.labelsize'     : 'medium',
    'ytick.labelsize'     : 'medium',
    'legend.fontsize'     : 'large',
})


fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(15, 5))

ax0.hist(df['area_sqmi'], 20, facecolor='green', alpha=0.75)
ax0.set_title('County Areas')
ax0.set_ylabel('number of counties')
ax0.set_xlabel('Area (in sq. miles)')

ax1.hist(df['pop_est_2015'], 20, facecolor='blue', alpha=0.75)
ax1.set_title('2015 Populations')
ax1.set_xlabel('Number of residents')

ax2.hist(df['pop_den_2015'], 20, facecolor='red', alpha=0.75)
ax2.set_title('2015 Population Densities')
ax2.set_xlabel('Population density (res./sq. mile)')

plt.savefig('../../figures/area_pop_dens_hist.png')
fig.tight_layout()
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(15, 5))

ax0.hist(df['log_area_sqmi'], 20, facecolor='green', alpha=0.75)
ax0.set_title('County Areas')
ax0.set_ylabel('number of counties')
ax0.set_xlabel('Area (in log10(sq. mile)')

ax1.hist(df['log_pop_est_2015'], 20, facecolor='blue', alpha=0.75)
ax1.set_title('2015 Populations')
ax1.set_xlabel('Log10 number of residents')

ax2.hist(df['log_pop_den_2015'], 20, facecolor='red', alpha=0.75)
ax2.set_title('2015 Population Densities')
ax2.set_xlabel('Pop. den. (in log10(res./sq. mile))')

plt.savefig('../../figures/logs_area_pop_dens_hist.png')
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 5))

ax0 = fig.add_subplot(1,1,1)
ax0.hist(df['pop_ch_10000_2015'], 20, facecolor='green', alpha=0.75)
ax0.set_title('2015 1-year population change')
ax0.set_ylabel('number of counties')
ax0.set_xlabel('Population change for every 10000 people')
fig.tight_layout()
```

```python
x = df['log_aland_sqmi']
y = df['log_pop_est_2015']
z = df['log_pop_den_2015']

fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(15, 5))

ax0.scatter(x, y, alpha=0.3);
ax0.set_title('Area - Population')
ax0.set_xlabel('Log of the area (in log(square miles)')
ax0.set_ylabel('Log of number of residents')

ax1.scatter(x, z, alpha=0.3);
ax1.set_title('Area - Population Density')
ax1.set_xlabel('Log of the area (in log(square miles)')
ax1.set_ylabel('Log of population density')

ax2.scatter(y, z, alpha=0.3);
ax2.set_title('Population - Population Density')
ax2.set_xlabel('Log of number of residents')
ax2.set_ylabel('Log of population density')

plt.savefig('../figures/area_pop_dens_scatters.png')
fig.tight_layout()
```
